```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

# A few kernel matrix in CNN


```python
identity_kernel=np.array([[0,0,0],
                          [0,1,0],
                          [0,0,0]])

sharpen_kernel=np.array([[0,-1,0],
                          [-1,5,-1],
                          [0,-1,0]])

edge_detection_kernel=np.array([[0,1,0],
                          [-1,8,1],
                          [-1,-1,-1,]])

box_blur_kernel=1/9*np.array([[1,1,1],
                              [1,1,1],
                              [1,1,1]])

laplacian_kernel = np.array([[0, 1, 0], 
                              [1, -4, 1], 
                              [0, 1, 0]])

emboss_kernel = np.array([[-2, -1, 0], 
                           [-1, 1, 1], 
                           [0, 1, 2]])

outline_kernel = np.array([[-1, -1, -1], 
                    [-1, 8, -1], 
                    [-1, -1, -1]])

bottom_sobel_kernel = np.array([[-1, -2, -1], 
                         [0, 0, 0], 
                         [1, 2, 1]])

right_sobel_kernel = np.array([[-1, 0, 1], 
                        [-2, 0, 2], 
                        [-1, 0, 1]])
```

# Read Image used to Matplotlib library

```python
import matplotlib.image as mpimg
image=mpimg.imread('2.jpg',0)
plt.imshow(image)
plt.show()
print(image.shape)
```

# Read Image used to Opencv library
<font size="3">Opencv represent RGB images as multi-dimensional Numpy array, but reverse order.
This means that the images are represent in BRG order than RGB. So we need reverse it!</font>

```python
image1=cv2.imread('3.jpg')
image1=cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
plt.imshow(image1)
# image2=cv2.imread('3.jpg')
# plt.imshow(image2)
plt.show()
```

# Convolution with kernel matrix (filter image)

```python
kernel_matrix=[identity_kernel,sharpen_kernel,edge_detection_kernel,box_blur_kernel,laplacian_kernel
              ,emboss_kernel,outline_kernel,bottom_sobel_kernel,right_sobel_kernel]
titles=['Identity','Sharpen','Edge Detection','Box Blur','Laplacian','Emboss',
        'Outline','Bottom Sobel','Right Sobel']

fix,axs=plt.subplots(3,3,figsize=(15,15))
for i in range(len(titles)):
    result=cv2.filter2D(image,-1,kernel_matrix[i])
    plt.subplot(3,3,i+1)
    plt.imshow(result)
    plt.title(titles[i]+',Shape:{}'.format(result.shape))
    
plt.show()
```
