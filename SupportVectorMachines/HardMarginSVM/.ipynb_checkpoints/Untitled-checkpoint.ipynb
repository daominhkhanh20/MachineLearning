```python
# for linearly separable data set
from cvxopt import solvers, matrix
import numpy as np
from matplotlib import pyplot as plt
from sklearn.svm import SVC  # Support vector Classification
np.random.seed(2)
```

# Generate Data
* N is number data points in each class
* cov is the covariance matrix


```python
means = [[2, 2], [4, 6]]
cov = [[.3, .2], [.2, .3]] 
N = 1000
x0 = np.random.multivariate_normal(means[0], cov, N)
x1 = np.random.multivariate_normal(means[1], cov, N)
# each colums is each data point
X = np.concatenate((x0.T, x1.T), axis=1)  # all data point
y_label = np.concatenate((np.ones((1, N)), -1 * np.ones((1, N))), axis=1)
```

```python
def display_result(w, b, x0, x1, X, S,title):
    x = np.linspace(-2, 6, 100)
    fig, ax = plt.subplots()
    w1, w2 = w[0, 0], w[1, 0]
    y = -w1 / w2 * x - b / w2
    y1 = -w1 / w2 * x - (b - 1) / w2
    y2 = -w1 / w2 * x - (b + 1) / w2
    plt.plot(x, y)
    plt.plot(x, y1, 'k')
    plt.plot(x, y2, 'k')

    plt.plot(x0[:, 0], x0[:, 1], 'bs', markersize=6, alpha=.8)
    plt.plot(x1[:, 0], x1[:, 1], 'ro', markersize=6, alpha=.8)
    
    if len(S) >0:
        for i in S:
            circle = plt.Circle((X[0, i], X[1, i]), 0.2, color='k', fill=False)
            ax.add_artist(circle)

    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Support Vector Machine with {}".format(title), fontsize=14)
    plt.savefig("{}.png".format(title))
    plt.show()
```

 # Quadratic programming problem
 $$ x= \underset{x}{\arg\min} \frac{1}{2} X^T K  X+q^T X+b$$

Condition: $$G X \leq h$$ <br>
           $$AX=b$$<br>
          
* K is a positive half-square matrix (satisfied the convex condition)
* G is matrix with shape(m*n)
* A is matrix with shape (p*n)
<br>
Loss function: $$\lambda=\underset{\lambda}{\arg\max}(\frac{-1}{2}\lambda^T K \lambda+\lambda)$$<br>
Condition: $$\lambda_i \geq 0, 0 \leq i \leq 2 n$$
           $$y^T\lambda=0$$
<br>

```python
def SVM_using_cvxopt_lib(x0, x1, X, y_label, N):
    # x= argmin 1/2*X.T*K*X+q.T*X+b
    V = np.concatenate((x0.T, -x1.T), axis=1)
    K = matrix(V.T.dot(V))
    q = matrix(-np.ones((2 * N, 1)))  # not transpose

    # build condition matrix

    # np.eye(n): create identity matrix with shape (n,n)
    # condition lamda i-th >=0  <==> -lamda <=0
    G = matrix(-np.eye(2 * N))  # the square matrix with shape (2n,2n)

    h = matrix(np.zeros((2 * N, 1)))
    A = matrix(y_label)
    b = matrix(np.zeros((1, 1)))
    solvers.options['show_progress'] = False
    sol = solvers.qp(K, q, G, h, A, b)
    lamda = np.array(sol['x'])

    epsilon = 1e-6
    # return array contains all of index i-th in lamda with lamda[i]>epsilon
    S = np.where((lamda > epsilon) == True)[0]
    V_s = V[:, S]
    X_s = X[:, S]
    y_s = y_label[:, S]
    lamda_s = lamda[S]
    w = V_s.dot(lamda_s)
    b = np.mean(y_s.T - w.T.dot(X_s))
    return (w, b, S)

(w1, b1, S1) = SVM_using_cvxopt_lib(x0, x1, X, y_label, N)
print(w1.T)
print(b1)
display_result(w1, b1, x0, x1, X,S1, "CVXOPT lib")
```

```python
 # Using sklearn library
def SVM_with_sklearn(X, y, N):
    X, y = X.T, y.reshape((2 * N,))  # each row is data point
    model = SVC(kernel='linear', C=1e5)
    model.fit(X, y)
    w = model.coef_
    b = model.intercept_
    return (w, b)

(w2, b2) = SVM_with_sklearn(X, y_label, N)
print(w2)
print(b2)

display_result(w2.T, b2, x0, x1, X,np.array([]), "Sklearn lib")
```
