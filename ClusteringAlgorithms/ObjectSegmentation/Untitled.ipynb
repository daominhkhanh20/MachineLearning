```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

img=mpimg.imread('/home/daominhkhanh/Desktop/MachineLearningBasic/ClusteringAlgorithms/ObjectSegmentation/girl.png')
plt.imshow(img)
plt.axis('off')
plt.show()

X=img.reshape((img.shape[0]*img.shape[1],img.shape[2]))

#each cluster is number color for picture
for cluster in [1,3,5,7,15,25,30]:
    kmeans=KMeans(n_clusters=cluster).fit(X)
    label=kmeans.predict(X)#assign label for each data point 
    
    img_temp=np.zeros_like(X)#generate array has the same shape X with all element =0
    #replace each pixel by its centroid
    for i in range(cluster):
        img_temp[label==i]=kmeans.cluster_centers_[i]
        
    img1=img_temp.reshape((img.shape[0],img.shape[1],img.shape[2]))
    #plt.imshow(img1,interpolation='nearest')
    plt.imshow(img1)
    plt.title("Cluste=%d"%cluster)
    plt.show()
 
    
```
