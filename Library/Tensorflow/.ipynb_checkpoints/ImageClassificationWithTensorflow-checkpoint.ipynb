```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
fashion_mnist=keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels)=fashion_mnist.load_data()
```

```python
train_images.shape
```

```python
print("Image for train:",len(train_images))
print("Image for test:",len(test_images))
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
              'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
#we have 10 class, each class is assigned 1 number(from 0 to 9)
print(train_labels)
```

```python
plt.imshow(train_images[0])
plt.colorbar()
```

```python
#scale these values to range 0 to 1
train_images=train_images/255.0
test_imgages=train_images/255.0

plt.figure(figsize=(10,10))
for i in range(36):
    plt.subplot(6,6,i+1)
    plt.xticks([])
    plt.yticks([])
#     plt.axis('off')
    plt.grid(False)
    plt.imshow(train_images[i],cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])

plt.show()
    
```

```python
#we build the neural network which has input layer, 1 hiden layer and output layer
model=keras.Sequential([
    #input layer
    #we reformat the image with 2 dimention into 1d-array with len(28*28)
    keras.layers.Flatten(input_shape=(28,28)),
    
    #hiden layer with 120 node in hiden layer, activation function: Rectified lineat unit
    keras.layers.Dense(120,activation='relu'),
    #ouput layer
    keras.layers.Dense(10)# 10 classes
])
```

# Complie Model 

Before the model is ready for training, it needs a few more settings.These are added during the model's compile step:

* Loss function:This measures how accurate the model is during training.You want to minimize this function to "steer" the model in the right direction.

* Optimizer:This is how the model is updated based on the data it sees and its loss function.

* Metrics:Used to monitor the training and testing steps. The following example uses accuracy, the fraction of the images that are correctly classified.

```python

model.compile(optimizer='adam',loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])
```

```python
print(type(train_images))
print(type(train_labels))
```

```python
#train model
model.fit(train_images,train_labels,epochs=20)
```

```python
test_loss,accuracy_test=model.evaluate(test_images,test_labels,verbose=2)
print("Accuracy for test:",accuracy_test)
```

```python
probability_model=keras.Sequential([
    model,
    keras.layers.Softmax()
])
predictions=probability_model(test_images)
```

```python
predictions[100]
```

```python
pre=np.argmax(predictions[100])
print("Predict class for image 100-th:",class_names[pre])
print("Correct class for image 100-th:",class_names[test_labels[100]])
```

```python
def plt_add_images(i,prediction_array,test_images,test_labels):
    true_label=test_labels[i]
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    predict_label=np.argmax(prediction_array)
    
    if true_label==predict_label:
        color='blue'
    else:
        color='red'
    
    plt.xlabel("{} {:0.2f}% ({})".format(class_names[predict_label],100*np.max(prediction_array),
                                       class_names[true_label]),color=color)

    
    
    
def plot_probability_array(i,prediction_array,test_labels):
    plt.yticks([])
    plt.xticks(range(10))
    true_label=test_labels[i]
    predict_label=np.argmax(prediction_array)
    
    my_graph=plt.bar(range(10),prediction_array,color='#777777')
    plt.ylim([0,1])
    my_graph[predict_label].set_color('red')
    my_graph[true_label].set_color('blue')
```

```python
n_rows=10
n_cols=4
n_images=n_rows*n_cols
plt.figure(figsize=(2*2*n_cols, 2*n_rows))
for i in range(n_images):
    plt.subplot(n_rows,2*n_cols,2*i+1)
    plt_add_images(i,predictions[i],test_images,test_labels)
    plt.subplot(n_rows,2*n_cols,2*i+2)
    plot_probability_array(i,predictions[i],test_labels)
    
plt.tight_layout()
plt.show()
```
