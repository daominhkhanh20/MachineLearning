# Reshape image 

```python
image.shape
```

```python
half=cv2.resize(image,(0,1),fx=0.5,fy=0.5)
bigger=cv2.resize(image,(300,400))
strech_near=cv2.resize(image,(700,1000),interpolation=cv2.INTER_NEAREST)
list_image=[image,half,bigger,strech_near]
titles=['original','half','bigger','Interpolation nearest']

for i in range(len(titles)):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.imshow(list_image[i])
    
plt.show()
```
